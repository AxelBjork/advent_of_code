```python
def parse_input(input_string):
    lines = input_string.strip().split('\n')
    directions = list(lines[0])
    mapping = {}
    for line in lines[2:]:
        key, values = line.split(' = ')
        mapping[key] = values.strip('()').split(', ')
    return directions, mapping


example_1 = """RL

AAA = (BBB, CCC)
BBB = (DDD, EEE)
CCC = (ZZZ, GGG)
DDD = (DDD, DDD)
EEE = (EEE, EEE)
GGG = (GGG, GGG)
ZZZ = (ZZZ, ZZZ)
"""

example_2 = """LLR

AAA = (BBB, BBB)
BBB = (AAA, ZZZ)
ZZZ = (ZZZ, ZZZ)
"""

example_3 = """LR

11A = (11B, XXX)
11B = (XXX, 11Z)
11Z = (11B, XXX)
22A = (22B, XXX)
22B = (22C, 22C)
22C = (22Z, 22Z)
22Z = (22B, 22B)
XXX = (XXX, XXX)
"""
direction_cache = {}
def get_next_direction(directions, mapping, start, min_steps=0):
    initial_start = start
    steps = 0
    dir_start = 0
    if initial_start in direction_cache.keys():
        start, steps, dir_start = direction_cache[initial_start]
        print(f'Resuming {initial_start} from {start} after {steps} steps, starting at {dir_start}')
    while steps < 100000000:
        for dir_id, direction in enumerate(directions):
            if dir_id < dir_start:
                continue
            dir_start = 0
            steps += 1
            if direction == 'L':
                start = mapping[start][0]
            elif direction == 'R':
                start = mapping[start][1]
            else:
                raise ValueError(f'Unknown direction: {direction}')
            if start[-1] == 'Z' and steps >= min_steps:
                print(f'Found solution for {initial_start} after {steps} steps, on {start} at {dir_id} dir count')
                # direction_cache[initial_start] = (start, steps, dir_id)
                return steps
    raise ValueError('Too many steps')

def get_starting_points(mapping):
    starting_points = {}
    for key in mapping.keys():
        if key[-1] == 'A':
            starting_points[key] = 0
    return starting_points

with open('adventofcode.com_2023_day_8_input.txt', 'r') as f:
    input_string = f.read()

content = parse_input(input_string)
# Find all steps which start with A
starting_points = get_starting_points(content[1])

min_steps = 0
while min_steps<10000000:
    for start_point in starting_points.keys():
        starting_points[start_point] = get_next_direction(*content, start_point, min_steps)
    # Check if all values in starting_points are equal
    if len(set(starting_points.values())) == 1:
        break
    # Set min_steps to the maximum value in starting_points
    min_steps = max(starting_points.values())
    print(f'New min_steps: {min_steps}, current starting_points: {starting_points}')

starting_points
print(set(starting_points.values()))
```

```python
import os

example_grid = """7 6 4 2 1
1 2 7 8 9
9 7 6 2 1
1 3 2 4 5
8 6 4 4 1
1 3 6 7 9"""

def get_data(real=False):
    if real:
        files = os.listdir(".")
        text_file = [file for file in files if ".txt" in file][0]
        with open(text_file, 'r') as f:
            input_string = f.read()
    else:
        input_string = example_grid
    input_grid = input_string.splitlines()
    print(f"Got {len(input_grid)} lines")
    return input_grid

# input_grid = get_data(True)
input_grid = get_data(True)
```

```python
def calc_safe(report_nums):
    if report_nums[0] - report_nums[1] > 0:
        report_nums.reverse()
    print(report_nums)
    for index in range(len(report_nums)-1):
        diff = report_nums[index+1] - report_nums[index]
        if 3 < diff or diff < 1:
            print(f"unsafe nums: {report_nums[index+1]}  {report_nums[index]}")
            return index
    return -1


def check_safe(input_grid):
    safe_reports = 0
    for line in input_grid:
        report_nums = [int(num) for num in line.split(" ")]
        if calc_safe(report_nums) == -1:
            safe_reports +=1
    return safe_reports

check_safe(input_grid)
```

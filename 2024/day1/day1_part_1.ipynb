```python
# The provided example grid
example_grid = """3   4
4   3
2   5
1   3
3   9
3   3"""
```

```python
import os

def get_data(real=False):
    if real:
        files = os.listdir(".")
        text_file = [file for file in files if ".txt" in file][0]
        with open(text_file, 'r') as f:
            input_string = f.read()
    else:
        input_string = example_grid
    input_grid = input_string.splitlines()
    print(f"Got {len(input_grid)} lines")
    return input_grid

input_grid = get_data(True)

```

```python

def parse_locations(in_data):
    loc_list_1 = [int(line.split("   ")[0]) for line in input_grid]
    loc_list_2 = [int(line.split("   ")[1]) for line in input_grid]

    loc_list_1.sort()
    loc_list_2.sort()

    print(loc_list_1)
    dist_sum = 0
    for i in range(len(loc_list_1)):
        dist_sum += abs(loc_list_1[i] - loc_list_2[i])
    print(dist_sum)
    return loc_list_1, loc_list_2




loc_list_1, loc_list_2 = parse_locations(input_grid)
```
